```python
from selenium import webdriver
from time import sleep
import pandas as pd
url = "https://www.census.gov/construction/bps/uspermits.html?fbclid=IwAR02DAW5FBC-At1R0g8qV_kFxrry7aIK5l1lfb--XDCLehPDUIlQ6r0lz8c"

# Driver setting
options = webdriver.ChromeOptions()
prefs = {'profile.default_content_settings.popups': 0, 'download.default_directory': '.'}
options.add_experimental_option('prefs', prefs)

#用chromedriver開啟Chrome
driver = webdriver.Chrome(chrome_options=options)
driver.get(url)

# Find excel and download
excel_data = driver.find_element_by_xpath("//a[@href='/construction/nrc/xls/permits_cust.xls']").click()

sleep(5)
driver.close()
```

```python
# Save into dataframe form , the number 2 implies "sheet 2"(where the Seasonally adjusted data saved)
df_SA = pd.read_excel('permits_cust.xls', 2)
# date
col2 = df_SA[df_SA.columns[0]]
# total value
col3 = df_SA[df_SA.columns[1]]

df_UA = pd.read_excel('permits_cust.xls', 1)
col1 = df_UA[df_UA.columns[1]]

# new list for date
date = []
# Seasonally Adjusted Annual Rate
SA_value = []
UA_value = []
# Data starts from row 8 to row 716 (1960-1-1 to 2019-1-1 = 12*59+1 = 709 = 716 - 8 + 1)
# Save into list
for i in range(8,717):
    date.append(col2[i])
    SA_value.append(col3[i])
    UA_value.append(col1[i+12])
# From list to dataframe
value_dict = {"Date":date, "PermitsUA" : UA_value,"PermitsSA": SA_value}
value_df = pd.DataFrame(value_dict)
value_df.set_index(keys= ["Date"],inplace=True)
# Finally we have
value_df.head(20)
```
